## Multiplexeur 4:1

Voici l'équation d'un multiplexeur 4:1 :

$Y = (D_0 \times \overline{S_1} \times \overline{S_0}) + (D_1 \times \overline{S_1} \times S_0) + (D_2 \times S_1 \times \overline{S_0}) + (D_3 \times S_1 \times S_0)$

Coder la fonction Python correspondante.

Nom: mux


Paramètres

* $S_0$ (bool/int) : Bit de sélection 0 (poids faible)
* $S_1$ (bool/int) : Bit de sélection 1 (poids fort)
* $D_0$ (bool/int) : Entrée de données 0 (sélectionnée quand $S_1$=0 et $S_0$=0)
* $D_1$ (bool/int) : Entrée de données 1 (sélectionnée quand $S_1$=0 et $S_0$=1)
* $D_2$ (bool/int) : Entrée de données 2 (sélectionnée quand $S_1$=1 et $S_0$=0)
* $D_3$ (bool/int) : Entrée de données 3 (sélectionnée quand $S_1$=1 et $S_0$=1)


Sortie: Y (int, 0 ou 1)

````python
def mux(S0, S1, D0, D1, D2, D3):
Y = (D0 and (not S1) and (not S0)) or (D1 and (not S1) and S0) or (D2 and S1 and (not S0)) or (D3 and S1 and S0)
 return 1 if Y else 0
````

Voici le code à utiliser pour tester votre fonction.

````python
assert mux(0, 0, 1, 0, 0, 0) == 1
assert mux(1, 0, 0, 1, 0, 0) == 1
assert mux(0, 1, 0, 0, 1, 0) == 1
assert mux(1, 1, 0, 0, 0, 1) == 1
print('All good ✌️')
````

## Additionneur 1 bit

On rappelle les équations logique de l'additionneur 1 bit:

Sortie: $S = A \oplus B \oplus C_{in}$

Retenue: $C_{out} = (A \times B) + (C_{in} \times (A \oplus B))$

 Coder une fonction python `adder` retournant $S$ et $C_{out}$ en fonction de $A$, $B$ et $C_{in}$.

````python
def adder(A, B, Cin):
    S = A ^ B ^ Cin
    Cout = (A and B) or (Cin and (A ^ B))
    return int(S), int(Cout)
````

````python
# Test
assert adder(0,0,0) == (0, 0)
assert adder(0,1,0) == (1, 0)
assert adder(0,0,1) == (1, 0)
assert adder(0,1,1) == (0, 1)
assert adder(1,1,1) == (1, 1)
````

## Additionneur n bits

Coder une fonction Python permettant d'additionner 2 mots de n bits stockés dans un tableau

Utiliser la fonction adder pour coder une fonction Python permettant d'additionner 2 mots de n bits stockés dans un tableau.

Exemple d'utilisation

```
num1 = [0,0,1,0]
num2 = [0,0,0,1]
solution = adder_n_bits(num1, num2)
```

> [0, 0, 1, 1]
def adder_n_bits(num1, num2):
    n = len(num1)
    result = [0] * n
    carry = 0
    for i in range(n - 1, -1, -1):
        S, carry = adder(num1[i], num2[i], carry)
        result[i] = S
    return [carry] + result

````python
# Test

num1 = [0,0,1,0]
num2 = [0,0,0,1]
solution = adder_n_bits(num1, num2)
print(solution)
assert solution == [0, 0, 0, 1, 1]
````

## Comparateur 4 bits

On considère un comparateur 4 bits avec les entrées $A = \{A_3, A_2, A_1, A_0\}$ et $B = \{B_3,B_2,B_1,B_0\}$, et 2 sorties $s_1$ et $s_0$ codées comme suit :
* $s_1s_0 = 00$ : Égalité ($A = B$)
* $s_1s_0 = 01$ : $A > B$
* $s_1s_0 = 10$ : $A < B$

Voici l'équation de $s_0$: 

$$
s_0 = (A_3 \overline{B_3}) +
       (\overline{A_3 \oplus B_3})(A_2 \overline{B_2}) +
       (\overline{A_3 \oplus B_3})(\overline{A_2 \oplus B_2})(A_1 \overline{B_1}) +
       (\overline{A_3 \oplus B_3})(\overline{A_2 \oplus B_2})(\overline{A_1 \oplus B_1})(A_0 \overline{B_0})
$$

Voici l'équation de $s_1$:

$$
s_1 = (\overline{A_3} B_3) +
       (\overline{A_3 \oplus B_3})(\overline{A_2} B_2) +
       (\overline{A_3 \oplus B_3})(\overline{A_2 \oplus B_2})(\overline{A_1} B_1) +
       (\overline{A_3 \oplus B_3})(\overline{A_2 \oplus B_2})(\overline{A_1 \oplus B_1})(\overline{A_0} B_0)
$$


Coder une fonction python implémentant cet algorithme.

**Conseil :** Certaines formules comme $\overline{A_3 \oplus B_3}$ sont souvent utilisées. Stockez les dans des variables.

````python
def comparator_4bit(A, B):
    A3,A2,A1,A0 = A
    B3,B2,B1,B0 = B
    eq3, eq2, eq1 = (A3==B3), (A2==B2), (A1==B1)
    s0 = (A3 and not B3) or (eq3 and A2 and not B2) or (eq3 and eq2 and A1 and not B1) or (eq3 and eq2 and eq1 and A0 and not B0)
    s1 = (not A3 and B3) or (eq3 and not A2 and B2) or (eq3 and eq2 and not A1 and B1) or (eq3 and eq2 and eq1 and not A0 and B0)
    return int(s1), int(s0)








````

````python
# Test
A = [1, 0, 1, 0]  # 5 en binaire (0101)
B = [0, 1, 1, 0]  # 6 en binaire (0110)
s1, s0 = comparator_4bit(A, B)
print(f"Sortie: {s1}{s0}") 
s1, s0 = comparator_4bit(B, A)
print(f"Sortie: {s1}{s0}")
s1, s0 = comparator_4bit(B, A)
print(f"Sortie: {s1}{s0}") 
````
