## Décalage de bits

Les décalages de bits (bit shifts) sont des opérations qui déplacent les bits d'un nombre dans une direction spécifique. Ils sont très utilisés en programmation bas niveau et pour les optimisations, car ils sont très rapides sur les processeurs.

Il existe deux types de décalage:

- Décalage à gauche de `n` bits  (`<< n`) :
    * Déplace tous les bits de `n` positions vers la gauche
    * Ajoute `n` 0 à droite
- Décalage à droite de `n` bits  (`>> n`) :
    * Déplace tous les bits de `n` positions vers la droite
    * Les `n` bits de droite sont perdus


**Exemple**

n = 10001

Décalage à gauche de 2 :

* Avant : 1 0 0 0 1
  
* Après : 1 0 0 0 1 0 0  (on ajoute deux zéros à droite)

Décalage à droite de 2 :

* Avant : 1 0 0 0 1
  
* Après :      1 0 0  (on perd les 2 derniers bits)

```python
value = 0b10001 # 17 en décimal (1×2⁴ + 0×2³ + 0×2² + 0×2¹ + 1×2⁰)
print(bin(value << 2)) # Décalage à gauche de 2 positions
print(bin(value >> 2)) # Décalage à droite de 2 positions
```

### Extraction du nième bit d'un entier

Pour extraire nième bit d'un entier, on effectue tout d'abord un décalage à droite de `n` bits. La valeur du bit recherché correspond au bit de poids faible. On fait donc un AND bit à bit avec 1 (`0b00..01`).

1. Valeur initiale :
   value = 17 (0b10001)

  
| 1  | 0  | 0  | 0  | 1  |  ← bits (de poids fort à faible)
   
 b4    b3    b2    b1    b0

2. Décalage à droite de n=2 positions :
   dec = value >> n
       = 0b10001 >> 2
       = 0b100 (qui vaut 4 en décimal)


| 1  | 0  | 0  | 0  | 1  |  ← valeur originale
   
      |
      v (décalage de 2 positions vers la droite)
   

| 1  | 0  | 0  |  ← résultat après décalage (les deux bits de droite sont perdus)


   Le bit qui était en position 2 (b2) se retrouve maintenant en position 0 (bit de poids faible)

2. Masquage avec AND pour isoler le bit de poids faible :
   bit = dec & 1
       = 0b100 & 0b001
       = 0b000
       = 0

| 1  | 0  | 0  |  ← résultat du décalage (0b100)

   
| 0  | 0  | 1  |  ← masque 0b1


| 0  | 0  | 0  |  ← résultat (le bit de poids faible est 0)


4. Résultat final :
   Le 2ème bit (en partant de 0) de 0b10001 (17) est 0.

   Représentation complète :
   value: 0b10001, dec: 0b100, Bit 2: 0


```python

value = 17
n = 2
dec = value >> n
bit = dec & 1
print(f"value: {bin(value)}, dec: {bin(dec)},  Bit {n}: {bit}")
```

### Inversion du nième bit d'un nombre

```python
value = 0b10100
n = 2
value = value ^ (1 << n)
print(bin(value))
```

### Création d'un masque de la bonne longueur

```python
value = 255
bit_length = value.bit_length()
print("number of bits needed to code the number: ", bit_length)
mask = (1 << bit_length) - 1
print(bin(mask))
```

## Opérations logiques bits à bits


- AND : `&`
- OR : `|`
- XOR : `^`
- NOT : `~`

Attention  l'opérateur `~` fonctionne avec des nombres signés. Par exemple, si on l'applique sur un nombre codé sur 4 bits, `0b0001`, nous n'obtiendrons pas la valeur souhaitée (`0b1110`) mais `-2` (`-0b10`) (le bit de signe a été également inversé) :

```python
value = 0b0001
print(bin(~value))
```

Il est donc préférable d'utiliser la valeur maximale que peut prendre le nombre. Par exemple, pour un mot de 4 bits: `0b1111`:

```python
value = 0b0001
print(bin(0b1111 ^ value))
```

Une adresse IP v4 est représentée comme un tableau de 4 entiers:

Coder le script Python permettant de retrouver les adresses de réseau et de broadcast en utilisant des opérateurs bit à bit.

```python
ip_address = [192, 168, 1, 1]
subnet_mask = [255, 255, 255, 0]
network_adress = []
for i in range (4) :
    network_adress.append(ip_address[i] & subnet_mask[i])
print(network_adress)

brodcast_address = []
for i in range (4) :
    brodcast_address.append( ip_address[i] |   255 ^ subnet_mask[i] )
print(brodcast_address)




```

### Algorithme de chiffrement symétrique

L'objectif est de coder un algorithme de chiffrement symétrique simple.

Une clé de chiffrement est définie sur 8 bits. Cette clé est symétrique: elle est utilisée pour coder et décoder chaque caractère du message.

Le codage et le décodage d'un caractère utilisent la méthode suivante : calcul du XOR bit à bit entre le code UTF8 du caractère et la clé.

Coder la fonction correspondante appelée `encode_decode_message` et prenant en paramètres le message et la clé de chiffrement.

```python
def encode_decode_message(message, key):
    encode_decode_message = ""
    for c in message:
        encode_decode_message += chr(ord(c) ^ key)
    return encode_decode_message
        
    




```

```python
# Test

message = "Ceci est un secret"


key = 0b11011011

print(message)

encoded_message = encode_decode_message(message, key)
print(encoded_message)

decoded_message = encode_decode_message(encoded_message, key)
print(decoded_message)

assert message == decoded_message
```
